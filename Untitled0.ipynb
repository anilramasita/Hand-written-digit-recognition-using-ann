```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import mnist
from keras.utils import to_categorical

# Step 1: Load and Preprocess the Data
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the input data (scaling the pixel values to the range [0, 1])
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Flatten the images from 28x28 to 784-dimensional vectors (needed for ANN)
x_train = x_train.reshape(-1, 28*28)  # -1 means it automatically infers the batch size
x_test = x_test.reshape(-1, 28*28)

# One-hot encode the labels (since there are 10 classes, digits 0-9)
y_train_encoded = to_categorical(y_train, 10)
y_test_encoded = to_categorical(y_test, 10)

# Step 2: Build the ANN Model
model = Sequential()

# Input layer with 784 units (28x28 pixels flattened) and hidden layer with 128 neurons
model.add(Dense(128, activation='relu', input_shape=(784,)))

# Output layer with 10 units (for digits 0-9) and softmax activation
model.add(Dense(10, activation='softmax'))

# Step 3: Compile the Model
# Use categorical crossentropy as the loss function and Adam optimizer
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Step 4: Train the Model
# Train the model on the training data (with validation on 20% of the training set)
history = model.fit(x_train, y_train_encoded, epochs=10, batch_size=32, validation_split=0.2)

# Step 5: Evaluate the Model
# Evaluate the model on the test data to see how well it performs
test_loss, test_acc = model.evaluate(x_test, y_test_encoded)
print(f"\nTest accuracy: {test_acc * 100:.2f}%")

# Step 6: Make Predictions
# Predict the first 5 test samples
predictions = model.predict(x_test[:5])

# Display the test images along with predicted and actual labels
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(f"Pred: {np.argmax(predictions[i])}\nActual: {y_test[i]}")
    plt.axis('off')
plt.show()

# Step 7: Test on a Specific Image
# You can test a specific image from the test set and see its prediction

# Let's choose the 10th image (index 9) for testing
n = 9  # You can change this to test different images
x_input = x_test[n].reshape(1, 28*28)

# Get the prediction for the selected image
prediction = model.predict(x_input)
predicted_digit = np.argmax(prediction[0])

# Display the selected test image and the predicted digit
plt.imshow(x_test[n].reshape(28, 28), cmap='gray')
plt.title(f"Predicted Digit: {predicted_digit}\nActual Digit: {y_test[n]}")
plt.axis('off')
plt.show()

# Optional: Print prediction probabilities
print(f"Prediction probabilities: {prediction[0]}")
```
